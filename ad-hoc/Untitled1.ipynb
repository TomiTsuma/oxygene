```python

def hash_string(input_string):
    # Convert the string to bytes
    input_bytes = input_string.encode('utf-8')

    # Create a SHA-256 hash object
    sha256_hash = hashlib.sha256()

    # Update the hash object with the bytes of the string
    sha256_hash.update(input_bytes)

    # Get the hexadecimal representation of the hash
    hex_digest = sha256_hash.hexdigest()

    return hex_digest

```

```python
hashed_string = hash_string(string)
print(f"SHA-256 hash: {hashed_string}")
```

```python
import zlib
import base64

def encode_paragraph(paragraph, max_length=30):
    # Compress the paragraph using zlib
    compressed_data = zlib.compress(paragraph.encode('utf-8'))

    # Encode the compressed data to Base64
    encoded_data = base64.urlsafe_b64encode(compressed_data)

    # Convert to a string and truncate if necessary
    encoded_string = encoded_data.decode('utf-8')


    return encoded_string

def decode_paragraph(encoded_string):
    # Decode the Base64 encoded string
    compressed_data = base64.urlsafe_b64decode(encoded_string)

    # Decompress the data
    original_paragraph = zlib.decompress(compressed_data).decode('utf-8')

    return original_paragraph

# Example usage
paragraph = "This is a sample paragraph that needs to be encoded to at most 30 characters."
encoded_paragraph = encode_paragraph(string)

print(f"Encoded: {encoded_paragraph}")
decoded_paragraph = decode_paragraph(encoded_paragraph)

print(f"Decoded: {decoded_paragraph}")
```
